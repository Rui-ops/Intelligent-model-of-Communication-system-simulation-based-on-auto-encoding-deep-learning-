```python
# importing libs
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.layers import Input, Dense, GaussianNoise
from keras.models import Model
from keras import regularizers
from keras.layers.normalization import BatchNormalization
from keras.optimizers import SGD
import random as rn
import os
```

```python
# defining parameters
M = 4 
k = np.log2(M)
k = int(k)
print ('M:',M,'k:',k)
```

```python
#generating data of size N
N = 100000
label = np.random.randint(M,size=N)
```

```python
# creating one hot encoded vectors
data = []
for i in label:
    temp = np.zeros(M)
    temp[i] = 1
    data.append(temp)
```

```python
data = np.array(data)
print (data.shape)
```

```python
temp_check = [17,23,45,67,89,96,72,250,350]
for i in temp_check:
    print(label[i],data[i])
```

```python
R = 4/8
n_channel = 4
print (int(k/R))
input_signal = Input(shape=(M,))
encoded = Dense(M, activation='relu')(input_signal)
enco=Dense(18, activation='relu')(encoded)
enco1=Dense(30, activation='relu')(enco)
enco2=Dense(16, activation='relu')(enco1)
encoded1 = Dense(n_channel, activation='linear')(enco2)
encoded2 = BatchNormalization()(encoded1)

EbNo_train = 5.01187 #  coverted 7 db of EbNo
encoded3 = GaussianNoise(np.sqrt(1/(2*R*EbNo_train)))(encoded2)

decoded = Dense(n_channel, activation='relu')(encoded3)
deco = Dense(16, activation="relu")(decoded)
deco1 = Dense(30, activation='relu')(deco)
deco2 = Dense(18, activation='relu')(deco1)
decoded1 = Dense(M, activation='softmax')(deco2)

autoencoder = Model(input_signal, decoded1)

autoencoder.compile(optimizer='adam', loss='categorical_crossentropy')
```

```python
#E2E Communication system
input_x = Input(shape=(M,))

x=input_x
for i in range(1,len(encoder.layers)):
    x = encoder.layers[i](x)
    
for j in range(1,len(channel.layers)):
    x = channel.layers[j](x)
    
for k in range(1,len(decoder.layers)):
    x = decoder.layers[k](x)

y = x
E2E=Model(input_x,y)

E2E.layers[-6].trainable = False
E2E.layers[-7].trainable = False
E2E.layers[-8].trainable = False
E2E.layers[-9].trainable = False
E2E.layers[-10].trainable = False
E2E.layers[-11].trainable = False

E2E.compile(optimizer='adam', loss='categorical_crossentropy')
print(E2E.summary())
```

```python
#E2E training
E2E.layers[-6].trainable = False
E2E.layers[-7].trainable = False
E2E.layers[-8].trainable = False
E2E.layers[-9].trainable = False
E2E.layers[-10].trainable = False
E2E.layers[-11].trainable = False

E2E.fit(data, data,
                epochs=50,
                batch_size=500,
                validation_data=(val_data, val_data))
```

```python
#Encoder
INP=Input(shape=(M,))
encoo=E2E.layers[1](INP)
encoo=E2E.layers[2](encoo)
encoo=E2E.layers[3](encoo)
encoo=E2E.layers[4](encoo)
encoo=E2E.layers[5](encoo)
encoo=E2E.layers[6](encoo)


Encoder=Model(INP, encoo)
Encoder.compile(optimizer='adam', loss='categorical_crossentropy')
Encoder.save('encoder.model')
print(Encoder.summary())
```

```python
#Decoder
INPU=Input(shape=(n_channel,))
decoo=E2E.layers[-5](INPU)
decoo=E2E.layers[-4](decoo)
decoo=E2E.layers[-3](decoo)
decoo=E2E.layers[-2](decoo)
decoo=E2E.layers[-1](decoo)


Decoder=Model(INPU, decoo)
Decoder.compile(optimizer='adam', loss='categorical_crossentropy')
Decoder.save('decoder.model')
print(Decoder.summary())
```
