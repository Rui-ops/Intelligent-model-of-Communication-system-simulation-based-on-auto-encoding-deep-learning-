```python
# importing libs
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.layers import Input, Dense, GaussianNoise
from keras.models import Model
from keras import regularizers
from keras.layers.normalization import BatchNormalization
from keras.optimizers import SGD
import random as rn
import os
from keras.models import load_model
```

```python
encoder = load_model('encoder.model')
decoder = load_model('decoder.model')
```

```python
# defining parameters
M = 4 
k = np.log2(M)
k = int(k)
print ('M:',M,'k:',k)
```

```python
#generating data of size N
N = 8192
label = np.random.randint(M,size=N)
```

```python
# creating one hot encoded vectors
data = []
for i in label:
    temp = np.zeros(M)
    temp[i] = 1
    data.append(temp)
```

```python
data = np.array(data)
print (data.shape)
```

```python
encoded = encoder.predict(data)
```

```python
def myprint(s):
    with open('deco_input.txt','w+') as f:
        print('\n')
        for i in range(0,s.size):
            print(s[i],file=f)
        
np.set_printoptions(threshold=np.inf)
myprint(encoded.flatten())
```

```python
# Loading data
deco_output=np.loadtxt('deco_output.txt')
#print(valoutput.shape)
Doutput=[]
for i in range(0,32768):
    Doutput.append(deco_output[i][1])
Doutput=np.array(Doutput)
Doutput=np.reshape(Doutput,(8192,4))
```

```python
N_val = 2048
val_label = np.random.randint(M,size=N_val)
val_data = []
for i in val_label:
    temp = np.zeros(M)
    temp[i] = 1
    val_data.append(temp)
val_data = np.array(val_data)
```

```python
Vinput=encoder.predict(val_data)
```

```python
def myprint1(s):
    with open('deco_vinput.txt','w+') as f:
        print('\n')
        for i in range(0,s.size):
            print(s[i],file=f)
        
np.set_printoptions(threshold=np.inf)
myprint1(Vinput.flatten())
```

```python
# Loading data
deco_voutput=np.loadtxt('deco_voutput.txt')
#print(valoutput.shape)
Dvoutput=[]
for i in range(0,8192):
    Dvoutput.append(deco_voutput[i][1])
Dvoutput=np.array(Dvoutput)
Dvoutput=np.reshape(Dvoutput,(2048,4))
```

```python
decoder.fit(Doutput, data,
                epochs=50,
                batch_size=500,
                validation_data=(Dvoutput, val_data))
```

```python
decoded = decoder.predict(Doutput)
pred_output = np.argmax(decoded,axis=1)
no_errors = (pred_output != label)
no_errors =  no_errors.astype(int).sum()
ber = no_errors 
print ('BER:',ber)
print(pred_output)
```
